```python
import numpy as np
import time

def task1_numpy(n):
    return np.diag(np.arange(1, n+1))

def task1_iter(n):
    A = []
    for i in range(n):
        row = []
        for j in range(n):
            row.append(i+1 if i == j else 0)
        A.append(row)
    return np.array(A, dtype=int)
```

```python
%%time
print(task1_iter(5), "\n")
```

## Завдання №3

```python
import numpy as np
import time

def task3_numpy(n, m, r, c):
    A = np.ones((n, m), dtype=int)
    A[::r, :] = 0
    A[:, ::c] = 0
    return A

def task3_iter(n, m, r, c):
    A = []
    for i in range(n):
        row = []
        for j in range(m):
            if i % r == 0 or j % c == 0:
                row.append(0)
            else:
                row.append(1)
        A.append(row)
    return np.array(A)
```

```python
%%time
print(task3_numpy(6, 7, 2, 3), "\n")
```

## Завдання №8

```python
import numpy as np
import time

def task8_numpy(arr):
    arr = np.array(arr)
    zeros = np.sum(arr == 0)
    nonzeros = np.sum(arr != 0)
    return zeros, nonzeros

def task8_iter(arr):
    zeros = sum(1 for x in arr if x == 0)
    nonzeros = sum(1 for x in arr if x != 0)
    return zeros, nonzeros
```

```python
%%time
arr8 = [3,4,0,9,8,2,4,0,8,4,1]
zeros, nonzeros = task8_numpy(arr8)
print(f"Нулів: {zeros}")
print(f"Не нулів: {nonzeros}")
```

## Завдання №11

```python
import numpy as np
import time

def task11_numpy(n):
    A = np.ones((n, n), dtype=int)
    B = np.zeros((n + 2, n + 2), dtype=int)
    B[1:-1, 1:-1] = A
    return B

def task11_iter(n):
    B = []
    for i in range(n + 2):
        row = []
        for j in range(n + 2):
            if i == 0 or i == n + 1 or j == 0 or j == n + 1:
                row.append(0)
            else:
                row.append(1)
        B.append(row)
    return B
```

```python
%%time
matrix = task11_numpy(4)
for row in matrix:
    print(row)
```

## Завдання №16

```python
import numpy as np
import time

def task16_numpy(n):
    A = np.arange(n + 1)
    mask = (A > n/4) & (A < 3*n/4)
    A[mask] = 0
    return A

def task16_iter(n):
    A = []
    for x in range(n + 1):
        if n/4 < x < 3*n/4:
            A.append(0)
        else:
            A.append(x)
    return A
```

```python
%%time
print("Task 16:", task16_iter(10), "\n")
```

## Завдання №18

```python
import numpy as np
import time

def task18_numpy(n):
    arr = np.random.randint(0, 99, size=n)
    unique_count = np.unique(arr).size
    return arr, unique_count

def task18_iter(n):
    arr = [np.random.randint(0, 99) for _ in range(n)]
    unique_count = len(set(arr))
    return arr, unique_count
```

```python
%%time
np.random.seed(151)
n = 30
arr, unique_count = task18_iter(n)
print("В послідовності", arr, "-", unique_count, "унікальних чисел")
```

## Завдання 2

```python
import numpy as np

np.set_printoptions(precision=8, suppress=True)

# Матриця A і вектор b (система 4x4)
# Система (в порядку рядків):
# x1  + 2x2  + 3x3 - 2x4 = 6
# x1  -  x2  - 2x3 - 3x4 = 8
# 3x1 + 2x2  -  x3 + 2x4 = 4
# 2x1 - 3x2  + 2x3 +  x4 = -8

A = np.array([
    [1,  2,  3, -2],
    [1, -1, -2, -3],
    [3,  2, -1,  2],
    [2, -3,  2,  1]
], dtype=float)

b = np.array([6, 8, 4, -8], dtype=float)

n = A.shape[0]
assert A.shape == (4,4) and b.shape == (4,)

print("Матриця A:")
print(A)
print("\nВектор b:")
print(b)

# Розв'язок за Крамером
detA = np.linalg.det(A)
print(f"\ndet(A) = {detA:.8f}")

if abs(detA) < 1e-12:
    raise ValueError("Матриця A сингулярна або близька до сингулярної: формули Крамера не застосовні.")

# Обчислюємо детермінанти замінних матриць
x_cramer = np.zeros(n, dtype=float)
det_list = []
for i in range(n):
    Ai = A.copy()
    Ai[:, i] = b            # замінюємо i стовпець на b
    detAi = np.linalg.det(Ai)
    det_list.append(detAi)
    x_cramer[i] = detAi / detA

print("\nДетермінанти для замінних матриць (det A_i):")
for i, d in enumerate(det_list, start=1):
    print(f" det A_{i} = {d:.8f}")
print("\nРозв'язок за формулами Крамера (x_cramer):")
print(x_cramer)

# Перевірка a матричним множенням A через крамер
Ax_cramer = A @ x_cramer
residual_cramer = Ax_cramer - b
print("\nПеревірка (A @ x_cramer):")
print("A @ x_cramer =", Ax_cramer)
print("Різниця (A @ x_cramer - b) =", residual_cramer)

# Перевірка b через обернену матрицю A^-1 через b
A_inv = np.linalg.inv(A)
x_inv = A_inv @ b
residual_inv = A @ x_inv - b

print("\nРозв'язок через обернену матрицю (x_inv = A^{-1} @ b):")
print(x_inv)
print("Різниця (A @ x_inv - b) =", residual_inv)

# Перевірка c numpy.linalg.solve
x_solve = np.linalg.solve(A, b)
residual_solve = A @ x_solve - b

print("\nРозв'язок через numpy.linalg.solve (x_solve):")
print(x_solve)
print("Різниця (A @ x_solve - b) =", residual_solve)

# Порівняння рішень через numpy.allclose()
print("\n- ПОРІВНЯННЯ -")
print("Cramer == inv :", np.allclose(x_cramer, x_inv))
print("Cramer == solve:", np.allclose(x_cramer, x_solve))
print("inv == solve  :", np.allclose(x_inv, x_solve))

# Чисельна оцінка
def vec_info(name, x):
    print(f"\n{name}:")
    print(x)
    print(" ||A@x - b|| (inf-norm) =", np.max(np.abs(A @ x - b)))
    print(" ||A@x - b|| (2-norm)   =", np.linalg.norm(A @ x - b))

vec_info("x_cramer", x_cramer)
vec_info("x_inv", x_inv)
vec_info("x_solve", x_solve)

# Округлимо рішення до зручного вигляду 8 знаків
print("\nОкруглені рішення (8 знаків):")
print("x_cramer :", np.round(x_cramer, 8))
print("x_inv    :", np.round(x_inv, 8))
print("x_solve  :", np.round(x_solve, 8))
```

## Завдання 3

```python
import numpy as np
import time
import pprint

# Матриці A і B
A_np = np.array([[4, 5, -2],
                 [3, -1, 0],
                 [-1, 0, 7]], dtype=float)

B_np = np.array([[2, 1, -1],
                 [0, 1, 3],
                 [5, 7, 3]], dtype=float)

# Те ж у вигляді списків списків для ітеративного підходу (доробити)
A_list = A_np.tolist()
B_list = B_np.tolist()

# Кількість повторів для вимірювання часу
REPS = 100000
# a NumPy-реалізація
def numpy_expr(A, B):
    # Обчислюємо 3A - (A - 2B)
    return 3*A - (A - 2*B)
    # (альтернативно: return 2*(A + B))

# b1 Ітеративно: через list comprehension
def listcomp_expr(A, B):
    # A та B - списки списків однакового розміру
    m = len(A)
    n = len(A[0])
    return [[int(3*A[i][j] - (A[i][j] - 2*B[i][j])) for j in range(n)] for i in range(m)]

# (b2) Ітеративно явні цикли
def loops_expr(A, B):
    m = len(A)
    n = len(A[0])
    # Нульова матриця результату
    R = [[0]*n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            # обчислення елемента
            R[i][j] = int(3*A[i][j] - (A[i][j] - 2*B[i][j]))
    return R

# Вимір часу
def time_func(fn, *args, reps=100000):
    # Викликаємо fn(*args) повторно reps разів і повертаємо час та останній результат
    t0 = time.perf_counter()
    res = None
    for _ in range(reps):
        res = fn(*args)
    t1 = time.perf_counter()
    return t1 - t0, res

# Виконання
# NumPy (перший виклик для розігріву за яким виключаємо одноразовый overhead)
_ = numpy_expr(A_np, B_np)

t_np, res_np = time_func(numpy_expr, A_np, B_np, reps=REPS)
# Перетворимо результат у цілі для наочного порівняння
res_np_int = res_np.astype(int)

# List comprehension
t_lc, res_lc = time_func(listcomp_expr, A_list, B_list, reps=REPS)

# Явні цикли
t_loops, res_loops = time_func(loops_expr, A_list, B_list, reps=REPS)
```

```python
print("Матриця A:")
pprint.pprint(A_list)
print("\nМатриця B:")
pprint.pprint(B_list)

print("\nРезультати обчислення (один прикладний результат)")
print("NumPy result (int):")
print(res_np_int)          # numpy array

print("\nList-comp result:")
pprint.pprint(res_lc)      # list of lists

print("\nLoops result:")
pprint.pprint(res_loops)   # list of lists

# Перевірки рівності через numpy.allclose
res_lc_np = np.array(res_lc, dtype=float)
res_loops_np = np.array(res_loops, dtype=float)

print("\nПеревірка рівності через numpy.allclose()")
print("NumPy vs List comprehension:", np.allclose(res_np_int, res_lc_np))
print("NumPy vs Loops             :", np.allclose(res_np_int, res_loops_np))
print("List comp vs Loops         :", np.allclose(res_lc_np, res_loops_np))

# Вимірювання часу
print("\nЧаси виконання (повний час, сек)")
print(f"NumPy ({REPS} повторів): {t_np:.6f} s  (середнє {t_np/REPS*1e6:.3f} μs/виклик)")
print(f"List comprehension ({REPS} повторів): {t_lc:.6f} s  (середнє {t_lc/REPS*1e6:.3f} μs/виклик)")
print(f"Loops ({REPS} повторів): {t_loops:.6f} s  (середнє {t_loops/REPS*1e6:.3f} μs/виклик)")
```
