## Завдання №2

Реалізувати **гном’яче сортування** для введеного списку елементів **від найбільшого до найменшого**.


```python
def gnome_sort():
    line = input("Введіть числа через пробіл: ").strip()
    if not line:
        print("Порожній ввід. Помилка!")
        return
    try:
        nums = [int(float(x)) if float(x).is_integer() else float(x) for x in line.split()]
    except ValueError:
        print("Вводьте лише числа. Для дробових використовуйте крапку. Помилка!")
        return
    i = 1
    while i < len(nums):
        if nums[i - 1] >= nums[i]:
            i += 1
        else:
            nums[i - 1], nums[i] = nums[i], nums[i - 1]
            if i > 1:
                i -= 1
    print("Відсортований список (від найбільшого до найменшого):", nums)

```
